```python
# Basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ML tools
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors

# Misc
import joblib
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Option A: upload from local computer
from google.colab import files
uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)   # change to read_excel if needed

# Quick checks
print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.describe().T
```

```python
# Drop duplicates
df = df.drop_duplicates().reset_index(drop=True)

# Fill numeric NaN with median
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Fill categorical NaN with "Unknown"
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna("Unknown")

print("After cleaning:", df.shape)
```

```python
features = ['danceability','energy','loudness','speechiness','acousticness',
            'instrumentalness','liveness','valence','tempo','duration_ms']

# Adjust if your dataset has slightly different names
print([f for f in features if f not in df.columns])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
X_scaled = pd.DataFrame(X_scaled, columns=features)
```

```python
X_scaled.hist(bins=30, figsize=(12,8))
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[features].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
```

```python
if 'playlist_genre' in df.columns:
    plt.figure(figsize=(12,4))
    df['playlist_genre'].value_counts().head(20).plot(kind='bar')
    plt.title("Top Genres")

    top_genres = df['playlist_genre'].value_counts().index[:6]
    sns.boxplot(x='playlist_genre', y='danceability',
                data=df[df['playlist_genre'].isin(top_genres)])
    plt.xticks(rotation=45)
    plt.title("Danceability by Genre")
```

```python
pca = PCA(n_components=2, random_state=42)
pca_components = pca.fit_transform(X_scaled)

df['pc1'] = pca_components[:,0]
df['pc2'] = pca_components[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(x='pc1', y='pc2', data=df, alpha=0.6)
plt.title("PCA Projection (no clusters yet)")
```

```python
inertia = []
sil = []
K = range(2,11)

for k in K:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertia.append(km.inertia_)
    sil.append(silhouette_score(X_scaled, labels))

plt.figure(figsize=(6,4))
plt.plot(K, inertia, 'o-')
plt.xlabel("k")
plt.ylabel("Inertia")
plt.title("Elbow Method")

for i in range(len(K)):
    print(f"k={K[i]}, silhouette={sil[i]:.3f}")
```

```python
k = 4  # <- change based on elbow/silhouette
kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
df['cluster'] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x='pc1', y='pc2', hue='cluster', palette='tab10', data=df, alpha=0.7)
plt.title(f"KMeans Clusters (k={k})")
```
